```python
from ultralytics import YOLO
import cv2

# Cargar modelo YOLO
model = YOLO("yolov8n.pt")

# Ruta de la imagen
ruta = r"C:\Users\Juan Camilo\Downloads\car.jpg"   # cambia aquí por el nombre REAL del archivo

# Cargar imagen
img = cv2.imread(ruta)

# Verifica si la imagen se cargó
if img is None:
    print("Error: No se pudo cargar la imagen. Verifica la ruta.")
    exit()

# Ejecutar detección
results = model(img)[0]

# Dibujar resultados
for box in results.boxes:
    x1, y1, x2, y2 = box.xyxy[0]
    cls = int(box.cls[0])
    label = model.names[cls]

    cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (0,255,0), 2)
    cv2.putText(img, label, (int(x1), int(y1)-5), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,0), 2)

# Mostrar imagen con detecciones
cv2.imshow("Detección YOLO", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2

# Cargar modelo YOLO pre-entrenado
model = YOLO("yolov8n.pt")

# Clases de vehículos que queremos detectar
vehiculos = ["car", "motorcycle", "bus", "truck"]

# Abrir webcam
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: No se pudo acceder a la cámara.")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Ejecutar detección
    results = model(frame)[0]

    # Contadores por tipo
    count_car = 0
    count_moto = 0
    count_bus = 0
    count_truck = 0

    # Contador general
    total_vehiculos = 0

    # Recorrer todas las detecciones
    for box in results.boxes:
        cls = int(box.cls[0])
        label = model.names[cls]

        if label in vehiculos:

            # Sumar a contador general
            total_vehiculos += 1

            # Contadores específicos
            if label == "car":
                count_car += 1
            elif label == "motorcycle":
                count_moto += 1
            elif label == "bus":
                count_bus += 1
            elif label == "truck":
                count_truck += 1

            # Dibujar cuadro y texto
            x1, y1, x2, y2 = box.xyxy[0]
            cv2.rectangle(
                frame,
                (int(x1), int(y1)),
                (int(x2), int(y2)),
                (0, 255, 0),
                2
            )
            cv2.putText(
                frame,
                label,
                (int(x1), int(y1) - 5),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.7,
                (0, 255, 0),
                2
            )

    # Mostrar todos los contadores en pantalla
    cv2.putText(frame, f"Carros: {count_car}", (20, 40),
                cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,255,255), 2)

    cv2.putText(frame, f"Motos: {count_moto}", (20, 80),
                cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,255,255), 2)

    cv2.putText(frame, f"Buses: {count_bus}", (20, 120),
                cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,255,255), 2)

    cv2.putText(frame, f"Camiones: {count_truck}", (20, 160),
                cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,255,255), 2)

    cv2.putText(frame, f"Total Vehículos: {total_vehiculos}", (20, 210),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 3)

    # Mostrar ventana
    cv2.imshow("Detección de Vehículos - YOLOv8", frame)

    # Salir con tecla q
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
